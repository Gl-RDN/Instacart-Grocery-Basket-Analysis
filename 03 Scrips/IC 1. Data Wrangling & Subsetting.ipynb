```python
# since we get a copy of the names we delete the first row
df_dep_t_new = df_dep_t[1:]
# [1:] means copy everything from the first row
```

```python
df_dep_t_new
```

```python
# assigning new name to a column using 'new_header' that copied earlier
df_dep_t_new.columns = new_header
```
