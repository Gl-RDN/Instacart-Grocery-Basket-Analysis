```python
# writing function called "price_label" that will apply a string label to every row within the dataframe
# designating it as a low-, mid-, or high-range product based on its price
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range Product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range Product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

## If Statement with For-Loops

```python
# creating new column "busiest day"
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

### 2. Create new column for 'Busiest Days' and 'Slowest Days'

```python
ords_prods_merge.head()
```

```python
# create new loop for 'busiest_days' 0,1 & 'slowest days' 3,4
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest days")
  elif value == 3 or value == 4:
    result.append('Least busy')
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# assigning the results to a 'busiest_days' column

ords_prods_merge['busiest_days'] = result
```

### 4. Identify busiest hours of the day. 

```python
# checking frequency of 'order_hour_of_day'

ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# create new loop for 'busiest_period_of_day' with 3 groups: Most-, Average-, Fewest - Orders
result = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9, 17, 8, 18]:
    result.append("Most Orders")
  elif value in [19, 20, 7, 21, 22, 23]:
    result.append("Average Orders")
  else:
    result.append("Fewest Orders")
```

```python
# assigning the results to a 'busiest_period_of_day' column

ords_prods_merge['busiest_period_of_day'] = result
```
